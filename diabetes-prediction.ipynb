# **Giới thiệu về Bộ Dữ Liệu**

Bộ dữ liệu chi tiết bao gồm thông tin sức khỏe và nhân khẩu học của 100.000 cá nhân, nhằm hỗ trợ nghiên cứu và mô hình dự đoán liên quan đến bệnh tiểu đường. Bộ dữ liệu này chứa các thông tin như: giới tính, tuổi, địa điểm, chủng tộc, tăng huyết áp, bệnh tim, tiền sử hút thuốc, chỉ số BMI, mức HbA1c, mức đường huyết và tình trạng tiểu đường.

**Ứng Dụng của Bộ Dữ Liệu**

Bộ dữ liệu này có thể được sử dụng trong nhiều mục đích phân tích và học máy, bao gồm:

Mô hình dự đoán: Xây dựng mô hình để dự đoán khả năng mắc bệnh tiểu đường dựa trên các đặc điểm nhân khẩu học và sức khỏe.

Phân tích sức khỏe: Nghiên cứu mối quan hệ giữa các chỉ số sức khỏe (ví dụ: BMI, mức HbA1c) với bệnh tiểu đường.

Nghiên cứu nhân khẩu học: Xem xét sự phân bố bệnh tiểu đường theo các nhóm dân số và địa điểm khác nhau.

Nghiên cứu y tế công cộng: Xác định các yếu tố nguy cơ của bệnh tiểu đường và đề xuất các biện pháp can thiệp cho nhóm có nguy cơ cao.

Nghiên cứu lâm sàng: Phân tích mối liên hệ giữa bệnh tiểu đường và các bệnh đi kèm như tăng huyết áp và bệnh tim.

**Các Phân Tích Tiềm Năng**

Thống kê mô tả: Tóm tắt bộ dữ liệu để hiểu rõ về xu hướng trung tâm và mức độ phân tán của các đặc trưng.

Phân tích tương quan: Xác định mối quan hệ giữa các đặc điểm trong bộ dữ liệu.

Mô hình phân loại: Sử dụng các thuật toán học máy để phân loại bệnh nhân mắc hoặc không mắc tiểu đường.

Phân tích xu hướng: Xem xét sự thay đổi của tỷ lệ mắc bệnh tiểu đường theo thời gian.

clinical_notes: Tổng hợp lâm sàng dựa trên các thuộc tính của bệnh nhân.

# III. Model baseline: Model: Logistic Regression

Train bằng Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
# Instantiate and train the logistic regression model
model_logistic = LogisticRegression(max_iter=5000)
model_logistic.fit(X_train_scaled, y_train)
```

Dự đoán và đánh giá model, ta sẽ xây dựng một hàm để dự đoán và đánh giá chất lượng của mô hình, chỉ số đánh giá ta sử dụng là f1_score 

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
def evaluate_model(model, X_test, y_test, verbose = True):
    """
    Hàm đánh giá mô hình với Accuracy và F1-score.
    
    Parameters:
        model: Mô hình đã huấn luyện (đã fit).
        X_test: Dữ liệu test (features).
        y_test: Nhãn thực tế của test set.
    
    Returns:
        None (in ra kết quả đánh giá).
    """
    # Dự đoán trên tập test
    y_pred = model.predict(X_test)
    # 🔹 Convert probabilities to binary labels (0 or 1)
    y_pred = (y_pred> 0.5).astype(int)  # Apply threshold
    # Tính Accuracy và F1-score, recall, precision
   
    f1 = f1_score(y_test, y_pred)
    # In ra kết quả
    if verbose:
       
        print(f'F1 score : {f1:.2f}')
    return  f1
```

```python
evaluate_model(model_logistic, X_test_scaled, y_test)
```

Ta sẽ thử cải thiện mô hình này theo các cách sau:

1. Giảm ngưỡng (threshold): Mô hình hiện tại đang dự đoán nếu xác suất>0.5 thì là 1, ta sẽ thử giảm xuống 0.4 để kiểm tra sự cải thiện của mô hình

```python
# Get probability scores for positive class
y_probs = model_logistic.predict_proba(X_test_scaled)[:, 1]
# Set a new threshold (e.g., 0.4 instead of 0.5)
new_threshold = 0.3
y_pred= (y_probs > new_threshold).astype(int)
# Evaluate again
# Tính f1_score
f1 = f1_score(y_test, y_pred)
# In ra kết quả

print(f'F1 score : {f1:.2f}')
```

Như vậy, ta thấy sau khi giảm threshold xuống 0.3, f1_score cũng giảm.

2. Sử dụng class_weights = 'balanced' để tăng sự tập trung của mô hình vào dự đoán lớp 1 (dương tính) 

```python
model_logistic = LogisticRegression(class_weight='balanced', max_iter=5000)
model_logistic.fit(X_train_scaled, y_train)
```

```python
evaluate_model(model_logistic, X_test_scaled, y_test)
```

F1_score giảm xuống nghiêm trọng

```python
# Try different weight values for class 1
weights = [{0: 1, 1: 2}, {0: 1, 1: 3}, {0: 1, 1: 5}]

for weight in weights:
    print(f"\nTesting class_weight = {weight}")

    # Train model with new class weights
    model_logistic = LogisticRegression(class_weight=weight, random_state=42)
    model_logistic.fit(X_train_scaled, y_train)
    # Evaluate the model
    evaluate_model(model_logistic, X_test_scaled, y_test)
    
```

Nhìn vào kết quả trên ta thấy việc dùng threshold 1:2 mang lại kết quả tốt nhất. Để cải thiện thêm nữa ta sẽ điều chỉnh về threshold bằng cách tăng threshold lên cao để hạn chế dương tính giả. 

```python
weights = {0: 1, 1: 2}
# Train model with new class weights
model_logistic = LogisticRegression(class_weight=weights, random_state=42)
model_logistic.fit(X_train_scaled, y_train)
# Get probability scores for positive class
y_probs = model_logistic.predict_proba(X_test_scaled)[:, 1]
# Set a new threshold (e.g., 0.6 instead of 0.5)
new_threshold = 0.7
y_pred= (y_probs > new_threshold).astype(int)
# Evaluate again
# Tính F1-score
f1 = f1_score(y_test, y_pred)
# In ra kết quả
print(f'F1 score : {f1:.2f}')
```

Mô hình LogisticRegression cuối cùng:

```python
# Instantiate and train the logistic regression model
model_logistic = LogisticRegression(max_iter=5000)
model_logistic.fit(X_train_scaled, y_train)
evaluate_model(model_logistic, X_test_scaled, y_test)
```

# IV. Lựa chọn đặc trưng 

Để tiếp tục cải tiến mô hình baseline (logistic Regression), ta sẽ thử lựa chọn đặc trưng. 

# 1. Chi-Square 

```python
import numpy as np
from sklearn.feature_selection import chi2
chi2_stats, p_values = chi2(X_train_scaled, y_train)
# Create a DataFrame to show results more clearly
import pandas as pd
chi2_results = pd.DataFrame({
    "Feature": X_train.columns,
    "Chi2 Score": chi2_stats,
    "p-value": p_values
})

# Sort features by importance (lowest p-value first)
chi2_results = chi2_results.sort_values(by="p-value")
chi2_results
# Chọn đặc trưng có p_values <0.05
chi_features = chi2_results [chi2_results['p-value'] <0.05]['Feature'].tolist()
print ('Các đặc trưng được lựa chọn là:')
print (chi_features)
print (f"Tổng số lượng đặc trưng được lựa chọn là {len(chi_features)}")
```

Dùng tệp dữ liệu này với mô hình baseline

```python
# Ensure X_train_scaled is a DataFrame with the same column names as X_train
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
# Ensure X_train_scaled is a DataFrame with the same column names as X_train
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)
# Select features based on p-value < 0.05
X_train_chi = X_train_scaled_df[chi_features]
X_test_chi = X_test_scaled_df[chi_features]
#Model Logistic
model_logistic.fit(X_train_chi, y_train)
#Evaluate the model:
evaluate_model(model_logistic, X_test_chi, y_test)
```

Như vậy ta nhận thấy việc sử dụng phương pháp thống kê Chi_square không giúp làm tăng f1_score của mô hình baseline, nhưng cũng giúp ta thấy rằng, dù ít hơn đặc trưng một cách đáng kể (21 so với 72 cột ban đầu), độ chính xác vẫn như vậy. 

# 2. ANOVA F-value for Feature Selection

```python
import numpy as np
from sklearn.feature_selection import f_classif
f_statistic, p_values = f_classif (X_train_scaled, y_train)
# Create a DataFrame to show results more clearly
import pandas as pd
f_results = pd.DataFrame({
    "Feature": X_train.columns,
    "Chi2 Score": f_statistic,
    "p-value": p_values
})

# Sort features by importance (lowest p-value first)
f_results  = f_results.sort_values(by="p-value")
f_results 
# Chọn đặc trưng có p_values <0.05
f_features = f_results [f_results ['p-value'] <0.05]['Feature'].tolist()
print ('Các đặc trưng được lựa chọn là:')
print (f_features)
print (f"Tổng số lượng đặc trưng được lựa chọn là {len(f_features)}")
```

Dùng tệp dữ liệu này với mô hình baseline

```python
# Select features based on p-value < 0.05
X_train_anova = X_train_scaled_df[f_features]
X_test_anova = X_test_scaled_df[f_features]
#Model Logistic
model_logistic.fit(X_train_anova, y_train)
#Evaluate the model:
evaluate_model(model_logistic, X_test_anova, y_test)
```

Độ chính xác cũng không thay đổi so với khi dùng toàn bộ tập dữ liệu 

# 3. Mutual Information for Feature Selection

```python
from sklearn.feature_selection import mutual_info_classif
# Compute Mutual Information
mi_scores = mutual_info_classif(X_train_scaled, y_train)
# Convert to DataFrame for better visualization
mi_results = pd.DataFrame({
    "Feature": X_train.columns,
    "MI Score": mi_scores
}).sort_values(by="MI Score", ascending=False)

# Print sorted Mutual Information scores
print(mi_results)
# Get selected feature indices
# Chọn các đặc trưng có MI Score > 0
threshold = 0
selected_features = mi_results[mi_results["MI Score"] > threshold]["Feature"].tolist()

print("\nSelected Features (MI > 0):", selected_features)
print (f"Tổng số lượng đặc trưng được lựa chọn là {len(selected_features)}")
```

Dùng tệp dữ liệu này với mô hình baseline

```python
X_train_mutual = X_train_scaled_df[selected_features]
X_test_mutual = X_test_scaled_df[selected_features]
#Model Logistic
model_logistic.fit(X_train_mutual, y_train)
#Evaluate the model:
evaluate_model(model_logistic, X_test_mutual, y_test)
```

# IX. VOTING

Voting là một kỹ thuật trong học máy tập hợp, trong đó nhiều mô hình (classifier) cùng dự đoán và kết quả cuối cùng được quyết định bằng cách tổng hợp các dự đoán của chúng. Có hai loại voting chính:

1️⃣ Hard Voting (Bỏ phiếu cứng)
Mỗi mô hình dự đoán một nhãn lớp (ví dụ: 0 hoặc 1).
Kết quả cuối cùng được xác định bằng bỏ phiếu đa số (lớp nào được nhiều mô hình dự đoán nhất sẽ thắng).

2️⃣ Soft Voting (Bỏ phiếu mềm)
Mỗi mô hình dự đoán xác suất của từng lớp thay vì nhãn lớp.
Kết quả cuối cùng dựa trên trung bình xác suất dự đoán của tất cả các mô hình.

Ở phương pháp này, ta cần chọn ra 3 mô hình để đưa vào, dựa vào kết quả tổng kết ở phần III. Xây dựng các mô hình Baseline, ta sẽ chọn Random forest, lightGBm và CatBoost
